```python
from pyspark.sql import SparkSession
```

```python
JOIN_COLUMN = 'TransactionID'
DATA_TYPE = 'train'

INPUT_FOLDER = './data/raw'
OUTPUT_FOLDER = './data/ingest'

INPUT_FILE_1 = f"{DATA_TYPE}_identity.csv"
INPUT_FILE_2 = f"{DATA_TYPE}_transaction.csv"
OUTPUT_FILE = f"{DATA_TYPE}_merged.csv"

FULL_INPUT_1 = f"{INPUT_FOLDER}/{INPUT_FILE_1}"
FULL_INPUT_2 = f"{INPUT_FOLDER}/{INPUT_FILE_2}"
FULL_OUTPUT = f"{OUTPUT_FOLDER}/{OUTPUT_FILE}"
```

```python
spark = SparkSession.builder \
    .appName("MergeLargeCSV") \
    .master("local[*]") \
    .getOrCreate()
```

```python
def merge_large_csv(input_path_1, input_path_2, output_path, join_column):
    print(f"Merging {input_path_1} and {input_path_2} into {output_path} using {join_column}")
    df1 = spark.read.csv(input_path_1, header=True, inferSchema=True)
    df2 = spark.read.csv(input_path_2, header=True, inferSchema=True)

    joined_df = df1.join(df2, on=join_column, how='inner')

    pandas_df = joined_df.toPandas()
    pandas_df.to_csv(output_path, index=False, encoding='utf-8')

    print("Merge completed.")
```

```python
merge_large_csv(FULL_INPUT_1, FULL_INPUT_2, FULL_OUTPUT, JOIN_COLUMN)
```
